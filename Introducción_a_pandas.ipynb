##Introducción a la libreria de pandas

```python
import pandas as pd 
import numpy as np
```

```python
mylist = [["Juanito", "Angel","Raza", "Edad"],
          ["Chato", "Exromántico", "Año", "Vida"],
          ["Migue", "Remi", "Manos", "Pies"],
          ["Juanita", "Mimí", "Libro", "Autor"],
          ["Tito", "Pérdida de tiempo", "Páginas", "Color"],
          ["Frifaier", "Come cerebro", "Miedo", "Espectativas"]]
```

```python
data_frame = pd.DataFrame(mylist)
print(data_frame)
```

```python
from pandas.core.indexes.datetimes import date_range
data_frame = pd.DataFrame(mylist, columns=["Nombre", "Especie", "Ruta", "Camino"])
print(data_frame)
```
